```python
# Question 1: Create a parent class Animal with a method speak() that prints a generic message. 
# Create a child class Dog that overrides the speak() method to print "Bark!".

class Animal:
    def speak(self):
        print("The animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Bark!")

# Testing the classes
print("Question 1 - Animal and Dog classes:")
animal = Animal()
dog = Dog()

animal.speak()  # Generic animal sound
dog.speak()     # Dog specific sound
```

```python
# Question 2: Write a program to create an abstract class Shape with a method area(). 
# Derive classes Circle and Rectangle from it and implement the area() method in both.

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width

# Testing the classes
print("\nQuestion 2 - Abstract Shape with Circle and Rectangle:")
circle = Circle(5)
rectangle = Rectangle(10, 6)

print(f"Circle area (radius=5): {circle.area():.2f}")
print(f"Rectangle area (10x6): {rectangle.area()}")
```

```python
# Question 3: Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. 
# Derive a class Car and further derive a class ElectricCar that adds a battery attribute.

class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type
    
    def display_info(self):
        print(f"Vehicle type: {self.type}")

class Car(Vehicle):
    def __init__(self, vehicle_type, brand, model):
        super().__init__(vehicle_type)
        self.brand = brand
        self.model = model
    
    def display_info(self):
        super().display_info()
        print(f"Brand: {self.brand}, Model: {self.model}")

class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, model, battery_capacity):
        super().__init__(vehicle_type, brand, model)
        self.battery = battery_capacity
    
    def display_info(self):
        super().display_info()
        print(f"Battery capacity: {self.battery} kWh")

# Testing the classes
print("\nQuestion 3 - Multi-level inheritance:")
my_electric_car = ElectricCar("Electric Vehicle", "Tesla", "Model 3", 75)
my_electric_car.display_info()
```

```python
# Question 4: Demonstrate polymorphism by creating a base class Bird with a method fly(). 
# Create two derived classes Sparrow and Penguin that override the fly() method.

class Bird:
    def fly(self):
        print("This bird can fly")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies swiftly through the air")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly, but it can swim excellently")

# Testing polymorphism
print("\nQuestion 4 - Polymorphism with Bird classes:")
birds = [Sparrow(), Penguin()]

for bird in birds:
    bird.fly()  # Each bird behaves differently
```

```python
# Question 5: Write a program to demonstrate encapsulation by creating a class BankAccount 
# with private attributes balance and methods to deposit, withdraw, and check balance.

class BankAccount:
    def __init__(self, account_holder, initial_balance=0):
        self.account_holder = account_holder
        self.__balance = initial_balance  # Private attribute
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive")
    
    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.__balance:
                self.__balance -= amount
                print(f"Withdrew ${amount}. New balance: ${self.__balance}")
            else:
                print("Insufficient funds")
        else:
            print("Withdrawal amount must be positive")
    
    def check_balance(self):
        return self.__balance

# Testing encapsulation
print("\nQuestion 5 - Encapsulation with BankAccount:")
my_account = BankAccount("John Doe", 1000)
print(f"Account holder: {my_account.account_holder}")
print(f"Initial balance: ${my_account.check_balance()}")

my_account.deposit(500)
my_account.withdraw(200)
print(f"Final balance: ${my_account.check_balance()}")
```

```python
# Question 6: Demonstrate runtime polymorphism using a method play() in a base class Instrument. 
# Derive classes Guitar and Piano that implement their own version of play().

class Instrument:
    def play(self):
        print("Playing a musical instrument")

class Guitar(Instrument):
    def play(self):
        print("Strumming the guitar strings - ðŸŽ¸")

class Piano(Instrument):
    def play(self):
        print("Playing beautiful melodies on piano keys - ðŸŽ¹")

# Testing runtime polymorphism
print("\nQuestion 6 - Runtime polymorphism with Instruments:")
instruments = [Guitar(), Piano(), Instrument()]

for instrument in instruments:
    instrument.play()  # Method resolved at runtime
```

```python
# Question 7: Create a class MathOperations with a class method add_numbers() to add two numbers 
# and a static method subtract_numbers() to subtract two numbers.

class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        result = a + b
        print(f"Addition: {a} + {b} = {result}")
        return result
    
    @staticmethod
    def subtract_numbers(a, b):
        result = a - b
        print(f"Subtraction: {a} - {b} = {result}")
        return result

# Testing class and static methods
print("\nQuestion 7 - MathOperations with class and static methods:")
MathOperations.add_numbers(15, 25)
MathOperations.subtract_numbers(50, 20)
```

```python
# Question 8: Implement a class Person with a class method to count the total number of persons created.

class Person:
    person_count = 0  # Class variable to keep track of instances
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        Person.person_count += 1
    
    @classmethod
    def get_person_count(cls):
        return cls.person_count
    
    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Testing person count
print("\nQuestion 8 - Person class with instance counter:")
print(f"Initial person count: {Person.get_person_count()}")

person1 = Person("Alice", 28)
person2 = Person("Bob", 35)
person3 = Person("Charlie", 42)

print(f"After creating 3 persons: {Person.get_person_count()}")
for i, person in enumerate([person1, person2, person3], 1):
    print(f"Person {i}: ", end="")
    person.display_info()
```

```python
# Question 9: Write a class Fraction with attributes numerator and denominator. 
# Override the str method to display the fraction as "numerator/denominator".

class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be zero")
        self.numerator = numerator
        self.denominator = denominator
    
    def __str__(self):
        return f"{self.numerator}/{self.denominator}"
    
    def simplify(self):
        def gcd(a, b):
            while b:
                a, b = b, a % b
            return a
        
        common_divisor = gcd(abs(self.numerator), abs(self.denominator))
        self.numerator //= common_divisor
        self.denominator //= common_divisor
        return self

# Testing Fraction class
print("\nQuestion 9 - Fraction class with __str__ override:")
fraction1 = Fraction(3, 4)
fraction2 = Fraction(8, 12)

print(f"Fraction 1: {fraction1}")
print(f"Fraction 2: {fraction2}")
print(f"Simplified Fraction 2: {fraction2.simplify()}")
```

```python
# Question 10: Demonstrate operator overloading by creating a class Vector 
# and overriding the add method to add two vectors.

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    def __str__(self):
        return f"Vector({self.x}, {self.y})"
    
    def __repr__(self):
        return self.__str__()

# Testing operator overloading
print("\nQuestion 10 - Operator overloading with Vector:")
vector1 = Vector(3, 4)
vector2 = Vector(1, 2)
result_vector = vector1 + vector2

print(f"Vector 1: {vector1}")
print(f"Vector 2: {vector2}")
print(f"Sum: {vector1} + {vector2} = {result_vector}")
```

```python
# Question 12: Implement a class Student with attributes name and grades. 
# Create a method average_grade() to compute the average of the grades.

class Student:
    def __init__(self, name, grades=None):
        self.name = name
        self.grades = grades if grades is not None else []
    
    def add_grade(self, grade):
        if 0 <= grade <= 100:
            self.grades.append(grade)
        else:
            print("Grade must be between 0 and 100")
    
    def average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)
    
    def display_info(self):
        avg = self.average_grade()
        print(f"Student: {self.name}")
        print(f"Grades: {self.grades}")
        print(f"Average Grade: {avg:.2f}")

# Testing Student class
print("\nQuestion 12 - Student class with grade average:")
student = Student("Michael", [85, 92, 78, 96, 88])
student.display_info()

# Adding more grades
student.add_grade(94)
print(f"\nAfter adding grade 94:")
student.display_info()
```

```python
# Question 13: Create a class Rectangle with methods set_dimensions() to set the dimensions 
# and area() to calculate the area.

class RectangleArea:
    def __init__(self, length=0, width=0):
        self.length = length
        self.width = width
    
    def set_dimensions(self, length, width):
        if length > 0 and width > 0:
            self.length = length
            self.width = width
            print(f"Dimensions set: {length} x {width}")
        else:
            print("Dimensions must be positive values")
    
    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2 * (self.length + self.width)
    
    def display_info(self):
        print(f"Rectangle: {self.length} x {self.width}")
        print(f"Area: {self.area()}")
        print(f"Perimeter: {self.perimeter()}")

# Testing Rectangle class
print("\nQuestion 13 - Rectangle class with dimensions and area:")
rect = RectangleArea()
rect.set_dimensions(12, 8)
rect.display_info()
```

```python
# Question 14: Create a class Employee with a method calculate_salary() that computes the salary 
# based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.

class Employee:
    def __init__(self, name, hourly_rate):
        self.name = name
        self.hourly_rate = hourly_rate
    
    def calculate_salary(self, hours_worked):
        return hours_worked * self.hourly_rate
    
    def display_info(self, hours_worked):
        salary = self.calculate_salary(hours_worked)
        print(f"Employee: {self.name}")
        print(f"Hourly Rate: ${self.hourly_rate}")
        print(f"Hours Worked: {hours_worked}")
        print(f"Salary: ${salary}")

class Manager(Employee):
    def __init__(self, name, hourly_rate, bonus):
        super().__init__(name, hourly_rate)
        self.bonus = bonus
    
    def calculate_salary(self, hours_worked):
        base_salary = super().calculate_salary(hours_worked)
        return base_salary + self.bonus
    
    def display_info(self, hours_worked):
        base_salary = super().calculate_salary(hours_worked)
        total_salary = self.calculate_salary(hours_worked)
        print(f"Manager: {self.name}")
        print(f"Hourly Rate: ${self.hourly_rate}")
        print(f"Hours Worked: {hours_worked}")
        print(f"Base Salary: ${base_salary}")
        print(f"Bonus: ${self.bonus}")
        print(f"Total Salary: ${total_salary}")

# Testing Employee and Manager classes
print("\nQuestion 14 - Employee and Manager salary calculation:")
employee = Employee("Sarah", 25)
manager = Manager("David", 35, 1000)

print("Employee Details:")
employee.display_info(40)

print("\nManager Details:")
manager.display_info(40)
```

```python
# Question 15: Create a class Product with attributes name, price, and quantity. 
# Implement a method total_price() that calculates the total price of the product.

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
    
    def total_price(self):
        return self.price * self.quantity
    
    def display_info(self):
        print(f"Product: {self.name}")
        print(f"Unit Price: ${self.price}")
        print(f"Quantity: {self.quantity}")
        print(f"Total Price: ${self.total_price()}")
    
    def update_quantity(self, new_quantity):
        self.quantity = new_quantity
        print(f"Quantity updated to {new_quantity}")

# Testing Product class
print("\nQuestion 15 - Product class with total price calculation:")
product1 = Product("Laptop", 899.99, 2)
product1.display_info()

print("\nAfter updating quantity:")
product1.update_quantity(3)
product1.display_info()
```

```python
# Question 16: Create a class Animal with an abstract method sound(). 
# Create two derived classes Cow and Sheep that implement the sound() method.

from abc import ABC, abstractmethod

class AnimalSound(ABC):
    def __init__(self, name):
        self.name = name
    
    @abstractmethod
    def sound(self):
        pass
    
    def introduce(self):
        print(f"This is a {self.name}")

class Cow(AnimalSound):
    def __init__(self):
        super().__init__("Cow")
    
    def sound(self):
        print("Moo! Moo!")

class Sheep(AnimalSound):
    def __init__(self):
        super().__init__("Sheep")
    
    def sound(self):
        print("Baa! Baa!")

# Testing abstract Animal class
print("\nQuestion 16 - Abstract Animal class with Cow and Sheep:")
animals = [Cow(), Sheep()]

for animal in animals:
    animal.introduce()
    animal.sound()
    print()
```

```python
# Question 17: Create a class Book with attributes title, author, and year_published. 
# Add a method get_book_info() that returns a formatted string with the book's details.

class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published
    
    def get_book_info(self):
        return f"'{self.title}' by {self.author}, published in {self.year_published}"
    
    def is_classic(self):
        current_year = 2025
        return (current_year - self.year_published) >= 50
    
    def display_details(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Year Published: {self.year_published}")
        print(f"Classic: {'Yes' if self.is_classic() else 'No'}")

# Testing Book class
print("\nQuestion 17 - Book class with formatted information:")
book1 = Book("To Kill a Mockingbird", "Harper Lee", 1960)
book2 = Book("The Great Gatsby", "F. Scott Fitzgerald", 1925)

print(book1.get_book_info())
print(book2.get_book_info())

print(f"\nDetailed info for {book1.title}:")
book1.display_details()
```

```python
# Question 18: Create a class House with attributes address and price. 
# Create a derived class Mansion that adds an attribute number_of_rooms.

class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price
    
    def display_info(self):
        print(f"Address: {self.address}")
        print(f"Price: ${self.price:,}")
    
    def calculate_property_tax(self, tax_rate=0.012):
        return self.price * tax_rate

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms
    
    def display_info(self):
        super().display_info()
        print(f"Number of Rooms: {self.number_of_rooms}")
    
    def calculate_luxury_tax(self):
        # Additional luxury tax for mansions
        if self.price > 1000000:
            return self.price * 0.005
        return 0
    
    def total_annual_cost(self):
        property_tax = self.calculate_property_tax()
        luxury_tax = self.calculate_luxury_tax()
        return property_tax + luxury_tax

# Testing House and Mansion classes
print("\nQuestion 18 - House and Mansion classes with inheritance:")

house = House("123 Oak Street", 350000)
mansion = Mansion("456 Elite Avenue", 2500000, 12)

print("Regular House:")
house.display_info()
print(f"Annual Property Tax: ${house.calculate_property_tax():,.2f}")

print("\nLuxury Mansion:")
mansion.display_info()
print(f"Annual Property Tax: ${mansion.calculate_property_tax():,.2f}")
print(f"Annual Luxury Tax: ${mansion.calculate_luxury_tax():,.2f}")
print(f"Total Annual Cost: ${mansion.total_annual_cost():,.2f}")
```
