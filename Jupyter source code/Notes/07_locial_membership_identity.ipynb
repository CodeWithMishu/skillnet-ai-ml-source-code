- The and operator returns the first operand if it is False, otherwise it returns the second operand.

```python
x = True
y = True

print(x and y) 

x = False
y = True

print(x and y) 

x = True
y = True

print(x or y)
```

(or) operator Return value True if one of the statement are true

```python
# or operator
x = 5
print(x>4 or x<4)
```

- The or operator returns the first operand if it is True, otherwise it returns the second operand.

```python
x = True
y = False

print(x or y)

x = False
y = True

print(x or y)

x = False
y = False

print((x or y))

```

- (not) Reverse the result ,return False if the result is True

```python
# not operator 
x = 9
print(not(x<6 and x>10))  
```

- The not operator evaluates to True if its operand is False, and False if its operand is True.

```python
x = True

print(not x)

x = False

print(not x) 
```

# Membership Operators:

- membership operators are used to test whether a value is a member of a sequence, such as a list, tuple, string, or set. There are two membership operators in Python: in and not in

```python
#in operators
string = "Hello, World!"
#print("Hello" not  in string) 
print("good" not in string)
```

```python
l=[6,7,8]
for i in l:
    print(i)
```

```python
#in operators
list = [1, 2, 3, 4, 5,7]

print(3 in list)
print(6 in list)
```

- not in operator: The not in operator returns True if a value is not found in the sequence, and False otherwise.

```python
a=[1,2,3]
print(1 in a)
```

```python
list = [1, 2, 3, 4, 5]

print(3 not in list) 
print(6 not in list) 

string = "Hello, World!"

print("Hello" not in string)
print("Goodbye" not in string)
```
