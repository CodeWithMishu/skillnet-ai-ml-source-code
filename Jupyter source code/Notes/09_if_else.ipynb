# Nested if-else:

- A nested if-else statement is a construct in which an if-else statement is placed inside another if-else statement. This allows you to test multiple conditions and execute different code blocks based on the outcome of those tests.

```python
x = 4
y = 2

if x > 6:  
    if y > 5:
        print("x and y are positive")
        
    else:
        print("x is positive but y is non-positive")
else:
    print("false")
    
```

```python
x=10
y=9
if x==y:
    print("x and y are equal")
if x>y:
    print("x is greater than y")
if y>x:
    print("y is greater then x")
```

# elif conditions:

- You can also use elif to include multiple conditions:

```python
#for example
x = 3

if x < 0:
    print("x is positive")

elif x>4:
    print("x is zero")
elif x >0:
    print("x is vhueu")
elif x <=0:
    print("x is zero")
else:
    print("x is negative")
```
