
- print is used to display the output of a function to the console. It is primarily used for debugging or checking the intermediate results of a function. The output of a print statement can be seen immediately and does not affect the flow of the program.

- return is used to specify the value that a function should return. The value returned by a function can be stored in a variable or used as an input to another function. The return statement terminates the function and returns the specified value to the caller. The returned value can be used for further processing and manipulation.

you should use print statements for debugging purposes, and use return statements to return the results of a function that can be used for further processing. If you only need to display the output of a function, use a print statement. If you need to return the results of a function for further processing, use a return statement.

```python
def square(x):
    result = x**2
    # print("Result inside function:", result)    
    return result

r = square(5)
r
print("Result outside function:", r)
```

- print statement inside the function displays the intermediate result, and the return statement returns the final result to the caller, which is displayed using another print statement.

# Recursion

- A function calling itself again and again to compute a value is referred to recursive function or recursion

```python
5!=5*4*3*2*1
```

```python
def fact(n):
    if n==1 or n==0:
        return 1
    else:
        return n*fact(n-1)
n=int(input("enter a number"))
r=fact(n)
print(r)
```

```python
#factorial of a number
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)
#                3*2*1

result = factorial(3)
print(result)
```

- Recursion is a powerful technique that can simplify the implementation of complex algorithms, but it is important to be mindful of the potential for infinite recursion, which can cause a stack overflow error

```python
#for example

def myfun():
    print("aditya mishra")
    myfun()
myfun() 
```

```python

def myfun():
    print("apple")
    myfun()
myfun()
```

MAP:

- functions returns a map object (which is an iterator) of the result after applying the given function to each item of a given iterable (list,tuple,etc)

```python
map(func_name, iter) 
```

```python
#syntax:

#map(func_name,iter)
def add(n):
    return n*n
number =[1,2,3,4]
result =map(add,number)
print(tuple(result))

# map(add,number)
```
